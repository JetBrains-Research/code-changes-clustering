```python
def set_exist():
    repos_dir = os.fsencode(repos_path)
    dirs = [os.fsdecode(el) for el in os.listdir(repos_dir) if os.path.isdir(repos_path + "/" + os.fsdecode(el))]
    return set(dirs)

set_exist()  
```
